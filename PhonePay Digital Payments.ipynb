#### Display Basic Statistics & data types for each dataset


```python
def show_stats(df, name):
    print(f"\n {name} (shape={df.shape}) ")
    print("Data types:")
    print(df.dtypes)
    print("\nNumeric summary:")
    desc=(df.describe().T)
    desc['median']=df.median(numeric_only=True)
    display(desc)
    print("\nTop values for object columns:")
    display(df.describe(include=['object']).T)

show_stats(state_txn_df, "State_Txn and Users")
show_stats(state_txnsplit_df, "State_TxnSplit")
show_stats(state_device_df, "State_DeviceData")
show_stats(district_txn_df, "District_Txn and Users")
show_stats(district_demo_df, "District Demographics")
```

#### Check for missing values

```python
def missing_report(df, name):
    miss = df.isnull().sum()
    miss = miss[miss > 0].sort_values(ascending=False)
    if miss.empty:
        print(f"{name}: No missing values.")
        return
    pct = (miss / len(df) * 100).round(2)
    report = pd.DataFrame({"missing_count": miss, "missing_pct": pct})
    print(f"\n{name} - missing columns:")
    display(report)
missing_report(state_txn_df, "State_Txn and Users")
missing_report(state_txnsplit_df, "State_TxnSplit")
missing_report(state_device_df, "State_DeviceData")
missing_report(district_txn_df, "District_Txn and Users")
missing_report(district_demo_df, "District Demographics")
```

#### Creating Summary of Number of states and Sttes with highest No.of districts

```python
def get_unique_states_and_districts(dfs):
    states = set()
    districts = set()
    for df in dfs:
        cols = [str(c).lower() for c in df.columns]
        for sc in ['state','state_name','statecode','state_name']:
            if sc in cols:
                states |= set(df[df.columns[cols.index(sc)]].dropna().unique())
                break
        for dc in ['district','district_name','districtcode','district_name']:
            if dc in cols:
                districts |= set(df[df.columns[cols.index(dc)]].dropna().unique())
                break
    return states, districts
states, districts = get_unique_states_and_districts([state_txn_df, district_txn_df, district_demo_df])
print("Unique states detected:", len(states))
print("Unique districts detected:", len(districts))
# Highest no.of districts.
if 'State' in district_demo_df.columns and 'District' in district_demo_df.columns:
    counts = district_demo_df['State'].value_counts()
    print("State with most districts (District Demographics):", counts.idxmax(), counts.max())
else:
    print("Could not detect 'State'/'District' columns exactly in District Demographics; inspect column names above.")
```

## Task 3: Data Quality Checks

#### Data consistency leavels across state & District levels

```python
tx_col = find_col(district_txn_df, ['transaction','txn','transactions'])
amt_col = find_col(district_txn_df, ['amount','amt'])
user_col = find_col(district_txn_df, ['registered','users','user'])
print("Detected district cols:", tx_col_d, amt_col_d, user_col_d)
print("Detected state cols (earlier):", tx_col, amt_col, )
if all([tx_col_d, amt_col_d, user_col_d, tx_col, amt_col]):
    district_agg = district_txn_df.groupby('State')[[tx_col_d, amt_col_d, user_col_d]].sum().reset_index().rename(
        columns={tx_col_d:'tx_district', amt_col_d:'amt_district', user_col_d:'user_district'}
    )
    state_agg = state_txn_df.groupby('State')[[tx_col, amt_col, user_col]].sum().reset_index().rename(
        columns={tx_col:'tx_state', amt_col:'amt_state', user_col:'user_state'}
    )
    merged = district_agg.merge(state_agg, on='State', how='outer')
    merged['tx_diff'] = merged['tx_district'] - merged['tx_state']
    merged['amt_diff'] = merged['amt_district'] - merged['amt_state']
    merged['user_diff'] = merged['user_district'] - merged['user_state']
    discrepancies = merged[(merged['tx_diff'].abs()>0) | (merged['amt_diff'].abs()>0) | (merged['user_diff'].abs()>0)]
    if discrepancies.empty:
        print("No discrepancies found between district sums and state-level numbers (for detected columns).")
    else:
        print("Discrepancies found (showing top 30):")
        display(discrepancies.head(30))
        discrepancies.to_csv("discrepancies_district_vs_state.csv", index=False)
        print("Saved: discrepancies_district_vs_state.csv")
else:
    print("Could not detect all required numeric columns to perform the consistency check. Inspect column names above.")
```

```python
user_col_state = find_col(state_txn_df, ['registered','users','registeredusers'])
pop_col_dem = find_col(district_demo_df, ['population','pop'])
print("Detected user_col_state:", user_col_state, "pop_col_dem:", pop_col_dem)
if user_col_state and pop_col_dem:
    pop_state = district_demo_df.groupby('State')[[pop_col_dem]].sum().reset_index().rename(columns={pop_col_dem:'Population'})
    users_state = state_txn_df.groupby('State')[[user_col_state]].sum().reset_index().rename(columns={user_col_state:'RegisteredUsers'})
    merged_users_pop = users_state.merge(pop_state, on='State', how='left')
    merged_users_pop['Users_to_Pop_Ratio'] = merged_users_pop['RegisteredUsers'] / merged_users_pop['Population']
    display(merged_users_pop.sort_values('Users_to_Pop_Ratio', ascending=False).head(20))
    merged_users_pop.to_csv("users_to_population_ratio.csv", index=False)
    print("Saved: users_to_population_ratio.csv")
    plt.figure(figsize=(12,4))
    plt.bar(merged_users_pop['State'], merged_users_pop['Users_to_Pop_Ratio'])
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig("users_to_population_ratio_by_state.png")
    plt.show()
    print("Saved: users_to_population_ratio_by_state.png")
else:
    print("Missing RegisteredUsers column in state data or Population column in district demographics.")
```

#### Correlate population density with transaction volume

```python
density_col = find_col(district_demo_df, ['density','population density','pop_density'])
tx_col_d = find_col(district_txn_df, ['transaction','txn','transactions'])
print("Detected density_col:", density_col, "tx_col_d:", tx_col_d)
if density_col and tx_col_d:
    d_left = district_txn_df.copy()
    d_right = district_demo_df.copy()
    merge_cols = ['State','District']
    if set(merge_cols).issubset(d_left.columns) and set(merge_cols).issubset(d_right.columns):
        merged_dd = d_left.merge(d_right, on=merge_cols, how='left')
    else:
        left_dcol = find_col(d_left, ['district','district_name'])
        right_dcol = find_col(d_right, ['district','district_name'])
        merged_dd = d_left.merge(d_right, left_on=left_dcol, right_on=right_dcol, how='left')
    merged_clean = merged_dd[[density_col, tx_col_d]].dropna()
    corr = merged_clean[density_col].corr(merged_clean[tx_col_d])
    print(f"Correlation density vs transaction volume (district-level): {corr:.4f}")
    merged_clean.to_csv("density_vs_tx_volume_dataset.csv", index=False)
    plt.figure(figsize=(6,5))
    plt.scatter(merged_clean[density_col], merged_clean[tx_col_d], alpha=0.5)
    plt.xlabel("Population density")
    plt.ylabel("Transaction volume")
    plt.title(f"Density vs Transaction volume (corr={corr:.3f})")
    plt.tight_layout()
    plt.savefig("density_vs_tx_volume_scatter.png")
    plt.show()
    print("Saved: density_vs_tx_volume_scatter.png and dataset CSV.")
else:
    print("Missing density or transaction columns needed for correlation.")
```
